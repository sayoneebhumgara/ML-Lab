```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Load the Diabetes dataset (regression example)
data = load_diabetes()
X = data.data
y = data.target
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# Create and train a Random Forest Regressor
regressor = RandomForestRegressor(n_estimators=100, random_state=42)
regressor.fit(X_train, y_train)
```

```python
# Make predictions
predictions = regressor.predict(X_test)
```

```python
# Calculate Mean Squared Error (MSE) and R-squared (R2) for evaluation
mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)
```

```python
print("Mean Squared Error:", mse)
print("R-squared (R2) Score:", r2)
```

```python
# Feature importances
feature_importances = regressor.feature_importances_
print("Feature Importances:", feature_importances)
```

```python
# Plot feature importances
plt.figure(figsize=(10, 6))
plt.barh(data.feature_names, feature_importances)
plt.xlabel('Feature Importance')
plt.ylabel('Features')
plt.title('Feature Importances (Random Forest Regressor)')
plt.show()
```

```python
# Plot true vs. predicted values with different colors and legend
plt.scatter(y_test[y_test < 100], predictions[y_test < 100], c='blue', label='True vs. Predicted (y < 100)', alpha=0.7)
plt.scatter(y_test[y_test >= 100], predictions[y_test >= 100], c='red', label='True vs. Predicted (y >= 100)', alpha=0.7)
plt.xlabel("True Values")
plt.ylabel("Predictions")
plt.title("True vs. Predicted Values (Random Forest Regressor)")
plt.legend()
plt.show()
```
